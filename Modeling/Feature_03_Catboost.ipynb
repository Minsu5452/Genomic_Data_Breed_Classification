```python
import pandas as pd
import random
import os
import numpy as np

from matplotlib import pyplot as plt

from catboost import CatBoostClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import f1_score

from sklearn.utils.class_weight import compute_class_weight
```

```python
class CFG:
    SEED = 42
```

```python
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
seed_everything(CFG.SEED) # Seed 고정
```

## Data Load

```python
X_train = pd.read_csv('/content/drive/MyDrive/유전체공모전/data/X_train_03.csv')
X_test = pd.read_csv('/content/drive/MyDrive/유전체공모전/data/X_test_03.csv')
y_train = pd.read_csv('/content/drive/MyDrive/유전체공모전/data/y_train_03.csv')
submit = pd.read_csv('/content/drive/MyDrive/유전체공모전/data/sample_submission.csv')
y_train = y_train['class']
```

```python
X_train = X_train.astype(str)
X_test = X_test.astype(str)
```

```python
Cat_list = X_train.columns.to_list()
```

### Model_Baseline

```python
model = CatBoostClassifier(objective='MultiClass', eval_metric="TotalF1:average=Macro", cat_features=Cat_list, class_weights={'A': 1, 'B': 0.5, 'C': 1.5}, task_type="GPU", random_state=CFG.SEED)
```

```python
model.fit(X_train, y_train)
```

```python
submit['class'] = model.predict(X_test)
```

```python
Best_sub = pd.read_csv('/content/drive/MyDrive/유전체공모전/submissions/baseline12.csv')
```

```python
submit.compare(Best_sub)
```
