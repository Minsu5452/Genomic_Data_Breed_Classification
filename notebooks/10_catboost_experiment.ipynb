## Import

```
import pandas as pd
import random
import os
import numpy as np

from matplotlib import pyplot as plt

from catboost import CatBoostClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import f1_score

from sklearn.utils.class_weight import compute_class_weight
```

```
class CFG:
    SEED = 42
```

```
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
seed_everything(CFG.SEED) # Seed 고정
```

## Data Load

```
X_train = pd.read_csv('data/X_train_03.csv')
X_test = pd.read_csv('data/X_test_03.csv')
y_train = pd.read_csv('data/y_train_03.csv')
submit = pd.read_csv('data/sample_submission.csv')
y_train = y_train['class']
```

```
X_train = X_train.astype(str)
X_test = X_test.astype(str)
```

```
Cat_list = X_train.columns.to_list()
```

### Model_Baseline

```
model = CatBoostClassifier(objective='MultiClass', eval_metric="TotalF1:average=Macro", cat_features=Cat_list, class_weights={'A': 1, 'B': 0.5, 'C': 1.5}, task_type="GPU", random_state=CFG.SEED)
```

```
model.fit(X_train, y_train)
```

```
submit['class'] = model.predict(X_test)
```

```
Best_sub = pd.read_csv('submissions/baseline12.csv')
```

```
submit.compare(Best_sub)
```
